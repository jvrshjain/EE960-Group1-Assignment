```python
#importing libraries

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
#Loading the data
smart_mon =pd.read_csv('Temp_and_humidity_dataset.csv')
```

```python
smart_mon.describe()
```

```python
smart_mon.info()
# No missing values in the data.
```

```python
X = smart_mon.iloc[:, 2].values
y = smart_mon.iloc[:, 1].values
```

```python
scalar = StandardScaler()
X = scalar.fit_transform(X.reshape(-1, 1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 5)
reg = LinearRegression()
reg.fit(X_train, y_train)
Ypred = reg.predict(X_test)

```

```python
y_train_predict = reg.predict(X_train)
rmse = np.sqrt(mean_squared_error(y_train, y_train_predict));
r2 = r2_score(y_train, y_train_predict)
mae = mean_absolute_error(y_train, y_train_predict)

print ("Train RMSE =", rmse)
print ("Train R2 Score  =", r2)
print ("Train MAE =", mae)
print ("\n")
```

```python
y_test_predict = reg.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_test_predict));
r2 = r2_score(y_test, y_test_predict)
mae = mean_absolute_error(y_test, y_test_predict)

print ("Test RMSE =", rmse)
print ("Test R2 Score  =", r2)
print ("Test MAE =", mae)
print ("\n")
```

```python
def plot_regressoin_line(X_train, y_train, X_test, y_test, reg):
    plt.figure(figsize=(12,6))
    plt.scatter(X_train, y_train, color='blue')
    plt.plot(X_train, reg.predict(X_train), color='red')
    plt.title('Temperature vs Humidity (Training set)')
    plt.xlabel('Temperature')
    plt.ylabel('Humidity')
    plt.show()

    plt.figure(figsize=(12,6))
    plt.scatter(X_test, y_test, color='blue')
    plt.plot(X_train, reg.predict
             (X_train), color='red')
    plt.title('Temperature vs Humidity (Test set)')
    plt.xlabel('Temperature')
    plt.ylabel('Humidity')
    plt.show()

plot_regressoin_line(X_train, y_train, X_test, y_test, reg)

```
